```python
# DO NOT EDIT THIS CELL

import copy

# Define the class for state
class State:
    """The fields are boolean values, they are true if the entity is in the left bank of the river and false if it is on the right"""

    def __init__(self, goat = "left", wolf = "left", cabbage = "left", farmer = "left"):
        self.goat = (goat == "left")
        self.farmer = (farmer == "left")
        self.cabbage = (cabbage == "left")
        self.wolf = (wolf == "left")

    def __eq__(self, other):
        """Define the equality operator"""
        return self.farmer == other.farmer and self.wolf == other.wolf and self.goat == other.goat and self.cabbage == other.cabbage

    def is_init_state(self):
        """Test whether this is the initial state: everybody in the left side"""
        return self.farmer and self.wolf and self.goat and self.cabbage

    def is_goal_state(self):
        """Test whether this is the goal state: everybody on the right side"""
        return (not self.farmer) and (not self.wolf) and (not self.goat) and (not self.cabbage)

    def is_danger_state(self):
        """Is this a dangerous state (wolf and goat or goat and cabbage together without farmer)"""
        if self.wolf == self.goat and self.wolf != self.farmer:
            return True
        if self.goat == self.cabbage and self.goat != self.farmer:
            return True
        return False

    def available_actions(self):
        actions = []
        if self.is_danger_state() or self.is_goal_state():
            return actions
        actions.append("farmer")
        if self.farmer == self.wolf:
            actions.append("wolf")
        if self.farmer == self.goat:
            actions.append("goat")
        if self.farmer == self.cabbage:
            actions.append("cabbage")
        return actions

    def transition(self, action):
        """action can be: "farmer" (farmer crosses river alone), "wolf" (farmer carries the wolf accross), "goat", "cabbage"
        Not all actions succeed: if the wolf was not on the same side of the river as the farmer, the "wolf" action is not possible.
        The function returns a new state.
        """
        if action not in self.available_actions():
            raise Exception("This action is not available in this state!")
        state = copy.copy(self)
        state.farmer = not state.farmer
        if action == "wolf":
            state.wolf = not self.wolf
        if action == "goat":
            state.goat = not self.goat
        if action == "cabbage":
            state.cabbage = not self.cabbage
        return state

    def __str__(self):
        str_left = f'{"farmer " if self.farmer else ""}{"wolf " if self.wolf else ""}{"goat " if self.goat else ""}{"cabbage" if self.cabbage else ""}{"none" if not (self.farmer or self.wolf or self.goat or self.cabbage) else ""}'
        str_right = f'{"farmer " if not self.farmer else ""}{"wolf " if not self.wolf else ""}{"goat " if not self.goat else ""}{"cabbage" if not self.cabbage else ""}{"none" if (self.farmer and self.wolf and self.goat and self.cabbage) else ""}'
        return "[" + str_left + " | " + str_right + "]"

    def __repr__(self):
        str_initial_state = f'{"initial state" if self.is_init_state() else "not initial state"}'
        str_danger_state = f'{"danger state" if self.is_danger_state() else "not danger state"}'
        str_goal_state = f'{"goal state" if self.is_goal_state() else "not goal state"}'
        str_actions = f'available actions = {self.available_actions()}'
        return "State:\n\t" + str(self) + "\n\t" + str_initial_state + "\n\t" + str_danger_state + "\n\t" + str_goal_state + "\n\t" + str_actions + "\n"
```

```python
# === Problem 2: Depth-First TREE Search


def dfs_tree_search(max_depth=20, verbose=True):
    """
    Pure tree search.
    - LIFO stack
    - Depth limit to prevent infinite descent
    - Backtrack pruning: don't apply the same action twice in a row
    Returns: list[str] actions to reach the goal, or None if not found.
    """
    start = State()
    goal_test = lambda s: s.is_goal_state()

    # stack holds tuples (state, path_actions, depth)
    stack = [(start, [], 0)]
    nodes_expanded = 0
    max_fringe = 1

    while stack:
        max_fringe = max(max_fringe, len(stack))
        state, path, depth = stack.pop()
        nodes_expanded += 1

        if goal_test(state):
            if verbose:
                print(f"Goal found. Nodes expanded: {nodes_expanded}, max fringe: {max_fringe}")
            return path

        if depth >= max_depth:
            continue  # stop diving deeper

        actions = state.available_actions()
        # push in reverse so the first listed action is explored first
        for a in reversed(actions):
            # simple local-cycle prune: avoid immediate undo (same action twice)
            # if path and a == path[-1]:
            #     continue
            next_state = state.transition(a)
            # optional guard: skip danger states (available_actions already helps)
            if next_state.is_danger_state():
                continue
            stack.append((next_state, path + [a], depth + 1))

    if verbose:
        print(f"No solution within depth {max_depth}. Nodes expanded: {nodes_expanded}, max fringe: {max_fringe}")
    return None


# ---- run DFS tree search for this puzzle assignment----
solution_actions = dfs_tree_search(max_depth=15, verbose=True)
print("DFS tree solution is :", solution_actions)

if solution_actions:
    s = State()
    print("Start:", s)
    for i, a in enumerate(solution_actions, 1):
        s = s.transition(a)
        print(f"Step {i}: {a:8} -> {s}")
    print("Goal reached..?", s.is_goal_state())
```

Implement breadth first tree search for this problem. Continue using your own representation from the depth first search problem.

```python
# Problem 3: Breadth First Tree Search

from collections import deque

class TreeNode:
    def __init__(self, state, action=None, parent=None):
        self.state = state
        self.action = action  # action that led to this state
        self.parent = parent
        self.children = []

    def get_path(self):
        path = []
        current = self
        while current is not None:
            if current.action is not None:
                path.append((current.action, current.state))
            current = current.parent
        return list(reversed(path))

    def add_child(self, state, action):
        child = TreeNode(state, action, self)
        self.children.append(child)
        return child

def breadth_first_search():
    """
    Implement breadth first tree search using a queue (FIFO)
    which returns the solution path if found, None otherwise
    """

    # Initialize
    initial_state = State()
    root = TreeNode(initial_state)
    fringe = deque([root])  # Queue for BFS (FIFO)
    nodes_expanded = 0
    max_fringe_size = 1

    print("Starting Breadth First Search...")
    print(f"Initial state: {initial_state}")

    while fringe:
        # Track maximum fringe size
        max_fringe_size = max(max_fringe_size, len(fringe))

        # Dequeue from front (queue behavior - FIFO)
        current_node = fringe.popleft()
        current_state = current_node.state
        nodes_expanded += 1

        print(f"\nExpanding node {nodes_expanded}: {current_state}")

        # Check if goal reached
        if current_state.is_goal_state():
            print(f"\nGoal found! Nodes expanded: {nodes_expanded}")
            print(f"Maximum fringe size: {max_fringe_size}")
            path = current_node.get_path()
            print(f"Solution path length: {len(path)}")
            print("\nSolution path:")
            print(f"Start: {initial_state}")
            for i, (action, state) in enumerate(path, 1):
                print(f"Step {i}: Action '{action}' -> {state}")
            return path

        # Generate successors
        available_actions = current_state.available_actions()
        print(f"Available actions: {available_actions}")

        # Add children to fringe (maintain order for consistent expansion)
        for action in available_actions:
            try:
                next_state = current_state.transition(action)
                child_node = current_node.add_child(next_state, action)
                fringe.append(child_node)  # Add to end for queue behavior
                print(f"  Added to fringe: action '{action}' -> {next_state}")
            except Exception as e:
                print(f"  Action '{action}' failed: {e}")

    print(f"\nNo solution found. Nodes expanded: {nodes_expanded}")
    print(f"Maximum fringe size: {max_fringe_size}")
    return None

# Run the breadth first search
print("="*60)
solution_path_bfs = breadth_first_search()
```

```python
# EDIT THIS PYTHON CELL
# write here the python implementation

# Problem 4: Graph Search for DFS and BFS

from collections import deque

# TreeNode class
class TreeNode:

    def __init__(self, state, action=None, parent=None):
        self.state = state
        self.action = action  # action that led to this state
        self.parent = parent
        self.children = []

    def get_path(self):

        path = []
        current = self
        while current is not None:
            if current.action is not None:
                path.append((current.action, current.state))
            current = current.parent
        return list(reversed(path))

    def add_child(self, state, action):

        child = TreeNode(state, action, self)
        self.children.append(child)
        return child

#Convert state to hashable tuple for closed set
def state_to_tuple(state):

    return (state.farmer, state.wolf, state.goat, state.cabbage)

# Depth first graph search with closed set implementation method
def depth_first_graph_search():

    initial_state = State()
    root = TreeNode(initial_state)
    fringe = [root]  # Stack for DFS
    closed_set = set()  # Set of visited states
    nodes_expanded = 0
    max_fringe_size = 1

    print("Starting Depth First Graph Search:: ")
    print(f"Initial state: {initial_state}")

    while fringe:
        max_fringe_size = max(max_fringe_size, len(fringe))

        # Pop from stack
        current_node = fringe.pop()
        current_state = current_node.state

        # Skip if already visited
        state_tuple = state_to_tuple(current_state)
        if state_tuple in closed_set:
            continue

        # Add to closed set
        closed_set.add(state_tuple)
        nodes_expanded += 1

        print(f"\nExpanding node {nodes_expanded}: {current_state}")

        # Checking if the goal is reached
        if current_state.is_goal_state():
            print(f"\nGoal found! Nodes expanded: {nodes_expanded}")
            print(f"Maximum fringe size: {max_fringe_size}")
            print(f"States in closed set: {len(closed_set)}")
            path = current_node.get_path()
            print(f"Solution path length: {len(path)}")
            print("\nSolution path:")
            print(f"Start: {initial_state}")
            for i, (action, state) in enumerate(path, 1):
                print(f"Step {i}: Action '{action}' -> {state}")
            return path

        available_actions = current_state.available_actions()
        print(f"Available actions are : {available_actions}")

        # Add children to fringe if not visited means
        for action in reversed(available_actions):
            try:
                next_state = current_state.transition(action)
                next_state_tuple = state_to_tuple(next_state)

                if next_state_tuple not in closed_set:
                    child_node = current_node.add_child(next_state, action)
                    fringe.append(child_node)
                    print(f"  Added to fringe: action '{action}' -> {next_state}")
                else:
                    print(f"  Skipped (already visited): action '{action}' -> {next_state}")
            except Exception as e:
                print(f"  Action '{action}' failed: {e}")

    print(f"\nNo solution found. Nodes expanded: {nodes_expanded}")
    return None

def breadth_first_graph_search():

    initial_state = State()
    root = TreeNode(initial_state)
    fringe = deque([root])  # Queue for BFS
    closed_set = set()  # Set of visited states
    nodes_expanded = 0
    max_fringe_size = 1

    print("Starting Breadth First Graph Search ::")
    print(f"Initial state: {initial_state}")

    while fringe:
        max_fringe_size = max(max_fringe_size, len(fringe))

        # Dequeue from front
        current_node = fringe.popleft()
        current_state = current_node.state

        # Skip if already visited
        state_tuple = state_to_tuple(current_state)
        if state_tuple in closed_set:
            continue

        # Add to closed set
        closed_set.add(state_tuple)
        nodes_expanded += 1

        print(f"\nExpanding node {nodes_expanded}: {current_state}")

        # Check if the goal is reached
        if current_state.is_goal_state():
            print(f"\nGoal found! Nodes expanded: {nodes_expanded}")
            print(f"Maximum fringe size: {max_fringe_size}")
            print(f"States in closed set: {len(closed_set)}")
            path = current_node.get_path()
            print(f"Solution path length: {len(path)}")
            print("\nSolution path:")
            print(f"Start: {initial_state}")
            for i, (action, state) in enumerate(path, 1):
                print(f"Step {i}: Action '{action}' -> {state}")
            return path

        available_actions = current_state.available_actions()
        print(f"Available actions: {available_actions}")

        # Add children to fringe if not visited
        for action in available_actions:
            try:
                next_state = current_state.transition(action)
                next_state_tuple = state_to_tuple(next_state)

                if next_state_tuple not in closed_set:
                    child_node = current_node.add_child(next_state, action)
                    fringe.append(child_node)
                    print(f"  Added to fringe: action '{action}' -> {next_state}")
                else:
                    print(f"  Skipped (already visited): action '{action}' -> {next_state}")
            except Exception as e:
                print(f"  Action '{action}' failed: {e}")

    print(f"\nNo solution found. Nodes expanded: {nodes_expanded}")
    return None

# Run both graph search algorithms
print("="*60)
print("GRAPH SEARCH ALGORITHMS")
print("="*60)

print("\n1. DEPTH FIRST GRAPH SEARCH:")
dfs_graph_solution = depth_first_graph_search()

print("\n" + "="*60)

print("\n2. BREADTH FIRST GRAPH SEARCH:")
bfs_graph_solution = breadth_first_graph_search()
```

```python
# Problem 7: Best-First Search - I have reused and modified from Problem 3 (BFS)
# Using the existing TreeNode class and BFS structure with priority queue

import heapq
from collections import deque

class TreeNode:

    def __init__(self, state, action=None, parent=None):
        self.state = state
        self.action = action  # action that led to this state
        self.parent = parent
        self.children = []
        self.heuristic_value = 0  # Added for best-first search

    def get_path(self):

        path = []
        current = self
        while current is not None:
            if current.action is not None:
                path.append((current.action, current.state))
            current = current.parent
        return list(reversed(path))

    def add_child(self, state, action):

        child = TreeNode(state, action, self)
        self.children.append(child)
        return child

    # Added for priority queue ordering
    def __lt__(self, other):
        return self.heuristic_value < other.heuristic_value

# Heuristic 1: Simple Manhattan Distance which count entities on the wrong side (should be on right side)
def heuristic_1_simple_distance(state):

    wrong_side_count = 0
    if state.farmer:  # farmer should be on right (False)
        wrong_side_count += 1
    if state.wolf:    # wolf should be on right (False)
        wrong_side_count += 1
    if state.goat:    # goat should be on right (False)
        wrong_side_count += 1
    if state.cabbage: # cabbage should be on right (False)
        wrong_side_count += 1
    return wrong_side_count


#Heuristic 2: Minimum Trips Estimate which is based on farmer position and remaining items to transport
def heuristic_2_minimum_trips(state):
    # Count items on left that need to be moved
    items_on_left = 0
    if state.wolf:
        items_on_left += 1
    if state.goat:
        items_on_left += 1
    if state.cabbage:
        items_on_left += 1

    if items_on_left == 0:
        # All items are on the right, farmer just needs to cross if on left
        return 1 if state.farmer else 0

    # Estimate trips: each item needs farmer to carry it across
    # Farmer needs to return for each additional item (except the last trip)
    if state.farmer:  # farmer on left with items
        return items_on_left * 2 - 1  # last trip doesn't need return
    else:  # farmer on right, needs to go back first
        return items_on_left * 2



#Best-First Search using Heuristic 1 used deque with heapq priority queue
def best_first_search_heuristic_1():
    initial_state = State()
    root = TreeNode(initial_state)
    root.heuristic_value = heuristic_1_simple_distance(initial_state)

    # Changed: Use priority queue instead of deque
    fringe = [root]  # Priority queue for best-first search
    heapq.heapify(fringe)

    visited_states = set()

    nodes_expanded = 0
    max_fringe_size = 1

    print("Starting Best-First Search with Heuristic 1 meethod (Simple Distance)...")
    print(f"Initial state: {initial_state}")
    print(f"Initial heuristic value: {root.heuristic_value}")

    while fringe:
        # Track maximum fringe size
        max_fringe_size = max(max_fringe_size, len(fringe))


        current_node = heapq.heappop(fringe)
        current_state = current_node.state

        #Check for cycles before expanding
        state_key = (current_state.farmer, current_state.wolf, current_state.goat, current_state.cabbage)
        if state_key in visited_states:
            continue
        visited_states.add(state_key)

        nodes_expanded += 1

        print(f"\nExpanding node {nodes_expanded}: {current_state}")
        print(f"Heuristic value: {current_node.heuristic_value}")

        # Checking ifthe goal is reached
        if current_state.is_goal_state():
            print(f"\nGoal found! Nodes expanded: {nodes_expanded}")
            print(f"Maximum fringe size: {max_fringe_size}")
            path = current_node.get_path()
            print(f"Solution path length: {len(path)}")
            print("\nSolution path:")
            print(f"Start: {initial_state}")
            for i, (action, state) in enumerate(path, 1):
                print(f"Step {i}: Action '{action}' -> {state}")
            return path


        available_actions = current_state.available_actions()
        print(f"Available actions: {available_actions}")


        for action in available_actions:
            try:
                next_state = current_state.transition(action)

                #Skip if next state already visited
                next_state_key = (next_state.farmer, next_state.wolf, next_state.goat, next_state.cabbage)
                if next_state_key in visited_states:
                    continue

                child_node = current_node.add_child(next_state, action)

                # Calculate heuristic value for the child
                child_node.heuristic_value = heuristic_1_simple_distance(next_state)

                # Use heappush instead of append
                heapq.heappush(fringe, child_node)
                print(f"  Added to fringe: action '{action}' -> {next_state} (h={child_node.heuristic_value})")
            except Exception as e:
                print(f"  Action '{action}' failed: {e}")

    print(f"\nNo solution found. Nodes expanded: {nodes_expanded}")
    print(f"Maximum fringe size: {max_fringe_size}")
    return None


#Best-First Search using Heuristic 2 (Minimum Trips)mehtod is replaced deque with heapq priority queue
def best_first_search_heuristic_2():

    initial_state = State()
    root = TreeNode(initial_state)
    root.heuristic_value = heuristic_2_minimum_trips(initial_state)


    fringe = [root]  # Priority queue for best-first search
    heapq.heapify(fringe)


    visited_states = set()

    nodes_expanded = 0
    max_fringe_size = 1

    print("Starting Best-First Search with Heuristic 2 (Minimum Trips)...")
    print(f"Initial state: {initial_state}")
    print(f"Initial heuristic value: {root.heuristic_value}")

    while fringe:
        # Track maximum fringe size
        max_fringe_size = max(max_fringe_size, len(fringe))


        current_node = heapq.heappop(fringe)
        current_state = current_node.state


        state_key = (current_state.farmer, current_state.wolf, current_state.goat, current_state.cabbage)
        if state_key in visited_states:
            continue
        visited_states.add(state_key)

        nodes_expanded += 1

        print(f"\nExpanding node {nodes_expanded}: {current_state}")
        print(f"Heuristic value: {current_node.heuristic_value}")


        if current_state.is_goal_state():
            print(f"\nGoal found! Nodes expanded: {nodes_expanded}")
            print(f"Maximum fringe size: {max_fringe_size}")
            path = current_node.get_path()
            print(f"Solution path length: {len(path)}")
            print("\nSolution path:")
            print(f"Start: {initial_state}")
            for i, (action, state) in enumerate(path, 1):
                print(f"Step {i}: Action '{action}' -> {state}")
            return path


        available_actions = current_state.available_actions()
        print(f"Available actions: {available_actions}")


        for action in available_actions:
            try:
                next_state = current_state.transition(action)


                next_state_key = (next_state.farmer, next_state.wolf, next_state.goat, next_state.cabbage)
                if next_state_key in visited_states:
                    continue

                child_node = current_node.add_child(next_state, action)

                child_node.heuristic_value = heuristic_2_minimum_trips(next_state)

                heapq.heappush(fringe, child_node)
                print(f"  Added to fringe: action '{action}' -> {next_state} (h={child_node.heuristic_value})")
            except Exception as e:
                print(f"  Action '{action}' failed: {e}")

    print(f"\nNo solution found. Nodes expanded: {nodes_expanded}")
    print(f"Maximum fringe size: {max_fringe_size}")
    return None

if __name__ == "__main__":
    print("="*80)
    print("PROBLEM 7: BEST-FIRST SEARCH IMPLEMENTATION")
    print("="*80)

    # Test the methods Heuristic 1 and 2
    print("\n" + "="*60)
    print("TESTING HEURISTIC 1: SIMPLE DISTANCE")
    print("="*60)
    solution_path_h1 = best_first_search_heuristic_1()

    print("\n" + "="*60)
    print("TESTING HEURISTIC 2: MINIMUM TRIPS")
    print("="*60)
    solution_path_h2 = best_first_search_heuristic_2()

    print("\n" + "="*60)
    print("COMPARISON DETAILS")
    print("="*60)

    if solution_path_h1:
        print(f"Heuristic 1 found solution with {len(solution_path_h1)} steps")
    else:
        print("Heuristic 1 did not find solution")

    if solution_path_h2:
        print(f"Heuristic 2 found solution with {len(solution_path_h2)} steps")
    else:
        print("Heuristic 2 did not find solution")

```

```python
# EDIT THIS PYTHON CELL
# write here the python implementation

# Problem 9: A* Tree Search Implementation and also i have reused the previous code for the implemnations
import heapq
from collections import deque

class TreeNode:

    def __init__(self, state, action=None, parent=None, path_cost=0):
        self.state = state
        self.action = action  # action that led to this state
        self.parent = parent
        self.children = []
        self.path_cost = path_cost  # g(n) - cost from start to this node
        self.heuristic_value = 0    # h(n) - heuristic estimate to goal
        self.f_value = 0            # f(n) = g(n) + h(n) - A* evaluation function

    def get_path(self):

        path = []
        current = self
        while current is not None:
            if current.action is not None:
                path.append((current.action, current.state))
            current = current.parent
        return list(reversed(path))

    def add_child(self, state, action, action_cost=1):

        child_path_cost = self.path_cost + action_cost
        child = TreeNode(state, action, self, child_path_cost)
        self.children.append(child)
        return child

    # A*: Compare by f-value instead of just heuristic
    def __lt__(self, other):
        if self.f_value == other.f_value:
            # Tie-breaking: prefer lower heuristic which is (closer to goal)
            return self.heuristic_value < other.heuristic_value
        return self.f_value < other.f_value


def heuristic_relaxed_transportation(state):

    items_to_transport = 0
    if state.wolf:      # wolf is on left, needs to go right
        items_to_transport += 1
    if state.goat:      # goat is on left, needs to go right
        items_to_transport += 1
    if state.cabbage:   # cabbage is on left, needs to go right
        items_to_transport += 1


    if items_to_transport == 0:
        return 1 if state.farmer else 0  # Only farmer might need to cross

    # Minimum actions needed in relaxed problem
    if state.farmer:  # farmer is on left with items
        return 2 * items_to_transport - 1
    else:  # farmer is on right, needs to go back first
        return 2 * items_to_transport


# ===== The A* TREE SEARCH IMPLEMENTATION =====

def a_tree_search_alg(max_nodes=1000, verbose=True):
    """
    A* Tree Search Implementation
    Modified from best-first search to use f(n) = g(n) + h(n)

    Key differences from best-first search:
    1. Uses f-value (g + h) for node selection instead of just h
    2. Tracks path cost g(n) from start to current node
    3. Guarantees optimal solution with admissible heuristic

    Args:
        max_nodes: Maximum number of nodes to expand
        verbose: Whether to print detailed search information

    Returns:
        Solution path (list of actions) or None if not found
    """

    initial_state = State()
    root = TreeNode(initial_state, path_cost=0)
    root.heuristic_value = heuristic_relaxed_transportation(initial_state)
    root.f_value = root.path_cost + root.heuristic_value  # f = g + h

    # Priority queue for A* (ordered by f-value)
    fringe = [root]
    heapq.heapify(fringe)

    visited_states = set()

    nodes_expanded = 0
    max_fringe_size = 1

    if verbose:
        print("="*60)
        print("A* TREE SEARCH WITH ADMISSIBLE HEURISTIC")
        print("="*60)
        print(f"Initial state: {initial_state}")
        print(f"Initial g(n): {root.path_cost}")
        print(f"Initial h(n): {root.heuristic_value}")
        print(f"Initial f(n): {root.f_value}")
        print()

    while fringe and nodes_expanded < max_nodes:
        # Track maximum fringe size
        max_fringe_size = max(max_fringe_size, len(fringe))

        # A* KEY: Select node with lowest f-value
        current_node = heapq.heappop(fringe)
        current_state = current_node.state

        # Cycle detection
        state_key = (current_state.farmer, current_state.wolf,
                    current_state.goat, current_state.cabbage)
        if state_key in visited_states:
            continue
        visited_states.add(state_key)

        nodes_expanded += 1

        if verbose:
            print(f"Expanding node {nodes_expanded}:")
            print(f"  State: {current_state}")
            print(f"  g(n) = {current_node.path_cost}")
            print(f"  h(n) = {current_node.heuristic_value}")
            print(f"  f(n) = {current_node.f_value}")

        # Goal test
        if current_state.is_goal_state():
            solution_path = [action for action, _ in current_node.get_path()]

            if verbose:
                print(f"\n OPTIMAL SOLUTION FOUND!")
                print(f"Nodes expanded: {nodes_expanded}")
                print(f"Maximum fringe size: {max_fringe_size}")
                print(f"Optimal cost: {current_node.path_cost}")
                print(f"Solution length: {len(solution_path)} steps")
                print()

                # Show complete solution path
                print("OPTIMAL SOLUTION PATH:")
                print("-" * 40)
                state = State()
                print(f"Start: {state}")
                for i, action in enumerate(solution_path, 1):
                    state = state.transition(action)
                    print(f"Step {i}: {action:20} -> {state}")
                print(f"\nFinal cost: {len(solution_path)} (optimal guaranteed)")

            return solution_path

        available_actions = current_state.available_actions()

        if verbose and available_actions:
            print(f"  Successors:")

        for action in available_actions:
            try:
                next_state = current_state.transition(action)
                next_state_key = (next_state.farmer, next_state.wolf,
                                next_state.goat, next_state.cabbage)

                # Skip if this is already visited
                if next_state_key in visited_states:
                    continue

                action_cost = 1  # Each action has cost 1
                child_node = current_node.add_child(next_state, action, action_cost)

                # Calculate A* values
                child_node.heuristic_value = heuristic_relaxed_transportation(next_state)
                child_node.f_value = child_node.path_cost + child_node.heuristic_value

                # Add to fringe
                heapq.heappush(fringe, child_node)

                if verbose:
                    print(f"    {action:15} -> g={child_node.path_cost}, "
                          f"h={child_node.heuristic_value}, f={child_node.f_value}")

            except Exception as e:
                if verbose:
                    print(f"    {action:15} -> FAILED: {e}")

        if verbose:
            print()

    # No solution found
    if verbose:
        if nodes_expanded >= max_nodes:
            print(f" Search terminated - reached max nodes ({max_nodes})")
        else:
            print(f" No solution found after expanding {nodes_expanded} nodes")
        print(f"Maximum fringe size: {max_fringe_size}")

    return None


if __name__ == "__main__":
    print("PROBLEM 9: A* TREE SEARCH IMPLEMENTATION")
    print("Using Problem 8's Admissible Heuristic")

    # Run A* search
    solution = a_tree_search_alg(max_nodes=100, verbose=True)

    if solution:
        print(f"\nA* found solution: {solution}")
        print(f"Solution is guaranteed optimal due to admissible heuristic!")
    else:
        print("\nNo solution found within node limit.")

```

```python
# EDIT THIS PYTHON CELL
# write here the python implementation

# Problem 10: A* Graph Search Implementation

import heapq
from collections import deque

class TreeNode:

    def __init__(self, state, action=None, parent=None, path_cost=0):
        self.state = state
        self.action = action  # action that led to this state
        self.parent = parent
        self.children = []
        self.path_cost = path_cost  # g(n) - cost from start to this node
        self.heuristic_value = 0    # h(n) - heuristic estimate to goal
        self.f_value = 0            # f(n) = g(n) + h(n) - A* evaluation function

    def get_path(self):

        path = []
        current = self
        while current is not None:
            if current.action is not None:
                path.append((current.action, current.state))
            current = current.parent
        return list(reversed(path))

    def add_child(self, state, action, action_cost=1):

        child_path_cost = self.path_cost + action_cost
        child = TreeNode(state, action, self, child_path_cost)
        self.children.append(child)
        return child

    # A*: Compare by f-value instead of just heuristic
    def __lt__(self, other):
        if self.f_value == other.f_value:
            # Tie-breaking: prefer lower heuristic (closer to goal)
            return self.heuristic_value < other.heuristic_value
        return self.f_value < other.f_value


def heuristic_relaxed_transportation(state):

    items_to_transport = 0
    if state.wolf:      # wolf is on left, needs to go right
        items_to_transport += 1
    if state.goat:      # goat is on left, needs to go right
        items_to_transport += 1
    if state.cabbage:   # cabbage is on left, needs to go right
        items_to_transport += 1

    if items_to_transport == 0:
        return 1 if state.farmer else 0  # Only farmer might need to cross

    # Minimum actions needed in relaxed problem
    if state.farmer:  # farmer is on left with items
        return 2 * items_to_transport - 1
    else:  # farmer is on right, needs to go back first
        return 2 * items_to_transport


def a_graph_search_alg(max_nodes=1000, verbose=True):
    """
    A* Graph Search Implementation
    Modified from A* Tree Searrrch

    KEY DIFFERENCES FROM TREE SEARCH:
    1. Maintains CLOSED list (explored set) of fully processed states
    2. Checks if successor is already in CLOSED with better path
    3. Updates existing nodes in OPEN if better path found
    4. Prevents redundant exploration of same state via different paths
    5. More memory efficient for problems with many duplicate states

    Args:
        max_nodes: Maximum number of nodes to expand
        verbose: Whether to print detailed search information

    Returns:
        Solution path (list of actions) or None if not found
    """

    initial_state = State()
    root = TreeNode(initial_state, path_cost=0)
    root.heuristic_value = heuristic_relaxed_transportation(initial_state)
    root.f_value = root.path_cost + root.heuristic_value  # f = g + h

    # GRAPH SEARCH DATA STRUCTURES:
    # OPEN list: nodes to be examined in this(fringe/frontier)
    open_list = [root]
    heapq.heapify(open_list)

    # Track nodes in open list by state for efficient lookup
    open_dict = {}  # state_key -> node
    initial_key = (initial_state.farmer, initial_state.wolf,
                   initial_state.goat, initial_state.cabbage)
    open_dict[initial_key] = root

    # CLOSED list: nodes already examined (explored set)
    closed_dict = {}  # state_key -> best_path_cost

    nodes_expanded = 0
    nodes_generated = 0
    max_open_size = 1

    if verbose:
        print("="*60)
        print("A* GRAPH SEARCH WITH ADMISSIBLE HEURISTIC")
        print("="*60)
        print(f"Initial state: {initial_state}")
        print(f"Initial g(n): {root.path_cost}")
        print(f"Initial h(n): {root.heuristic_value}")
        print(f"Initial f(n): {root.f_value}")
        print()

    while open_list and nodes_expanded < max_nodes:
        # Track maximum open list size
        max_open_size = max(max_open_size, len(open_list))

        # A*: Select node with lowest f-value from OPEN
        current_node = heapq.heappop(open_list)
        current_state = current_node.state

        # Create state key for lookups
        state_key = (current_state.farmer, current_state.wolf,
                    current_state.goat, current_state.cabbage)

        # Remove from open_dict since we're expanding it
        if state_key in open_dict:
            del open_dict[state_key]

        # GRAPH SEARCH: Move current node from OPEN to CLOSED
        closed_dict[state_key] = current_node.path_cost
        nodes_expanded += 1

        if verbose:
            print(f"Expanding node {nodes_expanded}:")
            print(f"  State: {current_state}")
            print(f"  g(n) = {current_node.path_cost}")
            print(f"  h(n) = {current_node.heuristic_value}")
            print(f"  f(n) = {current_node.f_value}")
            print(f"  Open list size: {len(open_list)}")
            print(f"  Closed list size: {len(closed_dict)}")

        # Goal test
        if current_state.is_goal_state():
            solution_path = [action for action, _ in current_node.get_path()]

            if verbose:
                print(f"\n OPTIMAL SOLUTION FOUND!")
                print(f"Nodes expanded: {nodes_expanded}")
                print(f"Nodes generated: {nodes_generated}")
                print(f"Maximum open list size: {max_open_size}")
                print(f"Final closed list size: {len(closed_dict)}")
                print(f"Optimal cost: {current_node.path_cost}")
                print(f"Solution length: {len(solution_path)} steps")
                print()

                print("OPTIMAL SOLUTION PATH:")
                print("-" * 40)
                state = State()
                print(f"Start: {state}")
                for i, action in enumerate(solution_path, 1):
                    state = state.transition(action)
                    print(f"Step {i}: {action:20} -> {state}")
                print(f"\nFinal cost: {len(solution_path)} (optimal guaranteed)")

            return solution_path

        available_actions = current_state.available_actions()

        if verbose and available_actions:
            print(f"  Generating successors:")

        for action in available_actions:
            try:
                next_state = current_state.transition(action)
                next_state_key = (next_state.farmer, next_state.wolf,
                                next_state.goat, next_state.cabbage)

                # Calculate costs for potential successor
                action_cost = 1
                new_path_cost = current_node.path_cost + action_cost
                nodes_generated += 1

                # GRAPH SEARCH LOGIC:

                # Case 1: State already in CLOSED with better or equal path
                if next_state_key in closed_dict:
                    if closed_dict[next_state_key] <= new_path_cost:
                        if verbose:
                            print(f"    {action:15} -> SKIP (in closed, g={closed_dict[next_state_key]} <= {new_path_cost})")
                        continue
                    else:
                        # Rare case: found better path to closed state
                        # In practice, with consistent heuristic, this shouldn't happen
                        if verbose:
                            print(f"    {action:15} -> REOPEN (better path g={new_path_cost} < {closed_dict[next_state_key]})")
                        del closed_dict[next_state_key]

                # Case 2: State already in OPEN
                if next_state_key in open_dict:
                    existing_node = open_dict[next_state_key]
                    if existing_node.path_cost <= new_path_cost:
                        if verbose:
                            print(f"    {action:15} -> SKIP (in open, g={existing_node.path_cost} <= {new_path_cost})")
                        continue
                    else:
                        # Found better path to existing open node
                        if verbose:
                            print(f"    {action:15} -> UPDATE (better path g={new_path_cost} < {existing_node.path_cost})")
                        # Remove old node from open_list (it will be filtered out)
                        # Update existing node with better path
                        existing_node.parent = current_node
                        existing_node.action = action
                        existing_node.path_cost = new_path_cost
                        existing_node.f_value = new_path_cost + existing_node.heuristic_value
                        # Re-add to heap with updated priority
                        heapq.heappush(open_list, existing_node)
                        continue

                # Case 3: New state - create and add to OPEN
                child_node = current_node.add_child(next_state, action, action_cost)
                child_node.heuristic_value = heuristic_relaxed_transportation(next_state)
                child_node.f_value = child_node.path_cost + child_node.heuristic_value

                # Add to open structures
                heapq.heappush(open_list, child_node)
                open_dict[next_state_key] = child_node

                if verbose:
                    print(f"    {action:15} -> ADD (g={child_node.path_cost}, "
                          f"h={child_node.heuristic_value}, f={child_node.f_value})")

            except Exception as e:
                if verbose:
                    print(f"    {action:15} -> FAILED: {e}")

        if verbose:
            print()

    # No solution found
    if verbose:
        if nodes_expanded >= max_nodes:
            print(f" Search terminated - reached max nodes ({max_nodes})")
        else:
            print(f" No solution found after expanding {nodes_expanded} nodes")
        print(f"Nodes expanded: {nodes_expanded}")
        print(f"Nodes generated: {nodes_generated}")
        print(f"Maximum open list size: {max_open_size}")
        print(f"Final closed list size: {len(closed_dict)}")

    return None

if __name__ == "__main__":
    print("PROBLEM 10: A* GRAPH SEARCH IMPLEMENTATION")
    print("Modified from Problem 9 (A* Tree Search)")

    # Run A* graph search
    solution = a_graph_search_alg(max_nodes=100, verbose=True)

    if solution:
        print(f"\nA* Graph Search found optimal solution: {solution}")
        print(f"Solution is guaranteed optimal due to admissible heuristic!")
    else:
        print("\nNo solution found within node limit.")

```
